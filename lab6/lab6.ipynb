```python
from scipy.stats import norm, t, chi2
```

```python
#  1.1:   t-
# :
confidence_level_a = 0.8
confidence_level_b = 0.4
confidence_level_c = 0.3
degrees_of_freedom_t = 5

# 1.1.1:    (N(0,1))
# a. P(X  [-a, a]) = 0.8
a_norm = norm.ppf(0.9, loc=0, scale=1)  # symmetrical interval for 80% probability

# b. P(X > b) = 0.4
b_norm = norm.ppf(1 - confidence_level_b)

# c. P(X < -c) + P(X > c) = 0.3
c_norm = norm.ppf(1 - confidence_level_c / 2)

# 1.1.2: T-  5   (T(5))
# a. P(X  [-a, a]) = 0.8
a_t = t.ppf(0.9, degrees_of_freedom_t)

# b. P(X > b) = 0.4
b_t = t.ppf(1 - confidence_level_b, degrees_of_freedom_t)

# c. P(X < -c) + P(X > c) = 0.3
c_t = t.ppf(1 - confidence_level_c / 2, degrees_of_freedom_t)
print('1.1')
{
    'norm': {'a': a_norm, 'b': b_norm, 'c': c_norm},
    't': {'a': a_t, 'b': b_t, 'c': c_t}
}
```

```python
#  1.2:     t-
#   (N(0,1))
# 1: P(X  [-0.1, 0.1])
gamma1_norm = norm.cdf(0.1) - norm.cdf(-0.1)

# 2: P(X > 0.7)
gamma2_norm = 1 - norm.cdf(0.7)

# 3: P(X < -0.4) + P(X > 0.4)
gamma3_norm = norm.cdf(-0.4) + (1 - norm.cdf(0.4))

# T-  5   (T(5))
# 1: P(X  [-0.1, 0.1])
gamma1_t = t.cdf(0.1, degrees_of_freedom_t) - t.cdf(-0.1, degrees_of_freedom_t)

# 2: P(X > 0.7)
gamma2_t = 1 - t.cdf(0.7, degrees_of_freedom_t)

# 3: P(X < -0.4) + P(X > 0.4)
gamma3_t = t.cdf(-0.4, degrees_of_freedom_t) + (1 - t.cdf(0.4, degrees_of_freedom_t))
print('1.2')
{
    'norm': {'gamma1': gamma1_norm, 'gamma2': gamma2_norm, 'gamma3': gamma3_norm},
    't': {'gamma1': gamma1_t, 'gamma2': gamma2_t, 'gamma3': gamma3_t}
}
```

```python
#  1.3: -   10   ((10))
# a. P(X > a) = 0.4
a_chi2 = chi2.ppf(1 - confidence_level_b, 10)

# b. P(X < b) = 0.2
b_chi2 = chi2.ppf(confidence_level_b / 2, 10)

# : P(X > 12)
gamma_chi2 = 1 - chi2.cdf(12, 10)
print('1.3')
{
    'chi2': {'a': a_chi2, 'b': b_chi2, 'gamma': gamma_chi2}
}
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


#  7
sigma = 0.3  #  
mu = 7       #  
N = 70       #   
gamma = 93

sample = stats.norm.rvs(loc=mu, scale=sigma, size=N)

print(sample)
```

```python
alpha = 1 - gamma / 100  #  

#  (a):     ( )
z = norm.ppf(1 - alpha / 2)
x_bar = np.mean(sample)  #  
epsilon_mean_known_sigma = z * sigma / np.sqrt(N)
ci_mean_known_sigma = (x_bar - epsilon_mean_known_sigma, x_bar + epsilon_mean_known_sigma)

#  (b):     ( )
s = np.std(sample, ddof=1)  #   
t_val = t.ppf(1 - alpha / 2, df=N-1)
ci_mean_unknown_sigma = (x_bar - t_val * s / np.sqrt(N), x_bar + t_val * s / np.sqrt(N))

#  (c):     (.. )
chi2_lower = chi2.ppf(alpha / 2, df=N-1)
chi2_upper = chi2.ppf(1 - alpha / 2, df=N-1)
ci_variance = ((N - 1) * s**2 / chi2_upper, (N - 1) * s**2 / chi2_lower)

#  
print(" :", x_bar)
print("   ( ):", epsilon_mean_known_sigma)
print("    ( ):", ci_mean_known_sigma)
print("    ( ):", ci_mean_unknown_sigma)
print("    (.. ):", ci_variance)
```

```python
#  4.     N1
epsilon_target = epsilon_mean_known_sigma / 2
N1_exact = (z * sigma / epsilon_target) ** 2
N1_min = int(np.ceil(N1_exact))

print(" N:", N)
print("   1 =", epsilon_target)
print("  N1:", N1_exact)
print("  N1   :", N1_min)
print(" N1 / N:", N1_exact / N)

```

```python
df = pd.read_csv('daily_weather_data.csv')
df
```

```python

#  5.1:      
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
selected_month = 6  # 
filtered_df = df[df['date'].dt.month == selected_month]
countries = filtered_df['country'].unique()[10:20]  #   10 
filtered_df = filtered_df[filtered_df['country'].isin(countries)]

filtered_df
```

```python
from scipy import stats as st

#  5.2:      
gamma = 0.95  #  
alpha_mean = 1 - gamma

list_statistics = (
    filtered_df.groupby("country")["tavg"]
    .agg(["mean", "std", "count"])
    .rename(columns={"mean": "mean_tavg", "std": "std_tavg"})
)

#       
list_statistics = list_statistics[list_statistics["count"] > 1].copy()

#   
list_statistics["t_multiplier"] = list_statistics["count"].apply(
    lambda n: st.t.ppf((1 + gamma) / 2, df=n - 1)
)
list_statistics["epsilon_mean"] = (
    list_statistics["t_multiplier"] * list_statistics["std_tavg"] / np.sqrt(list_statistics["count"])
)
list_statistics["ci_mean_lower"] = list_statistics["mean_tavg"] - list_statistics["epsilon_mean"]
list_statistics["ci_mean_upper"] = list_statistics["mean_tavg"] + list_statistics["epsilon_mean"]

#   
degrees_of_freedom = list_statistics["count"] - 1
list_statistics["variance_tavg"] = list_statistics["std_tavg"] ** 2
chi2_lower = chi2.ppf(alpha_mean / 2, degrees_of_freedom)
chi2_upper = chi2.ppf(1 - alpha_mean / 2, degrees_of_freedom)
list_statistics["ci_var_lower"] = (
    degrees_of_freedom * list_statistics["variance_tavg"] / chi2_upper
)
list_statistics["ci_var_upper"] = (
    degrees_of_freedom * list_statistics["variance_tavg"] / chi2_lower
)

list_statistics[[
    "mean_tavg", "std_tavg", "count",
    "ci_mean_lower", "ci_mean_upper",
    "variance_tavg", "ci_var_lower", "ci_var_upper"
]]
```

```python

#  5.3:      errorbar
fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

axes[0].errorbar(
    list_statistics.index,
    list_statistics['mean_tavg'],
    yerr=list_statistics['epsilon_mean'],
    fmt='o', capsize=6, label='95% CI for mean'
)
axes[0].set_ylabel("Mean tavg")
axes[0].set_title("    tavg ()")
axes[0].grid(True, linestyle='--', color='gray', alpha=0.7)
axes[0].legend()

variance_yerr = np.vstack([
    list_statistics['variance_tavg'] - list_statistics['ci_var_lower'],
    list_statistics['ci_var_upper'] - list_statistics['variance_tavg']
])
axes[1].errorbar(
    list_statistics.index,
    list_statistics['variance_tavg'],
    yerr=variance_yerr,
    fmt='s', capsize=6, color='tab:red', label='95% CI for variance'
)
axes[1].set_xlabel("Country")
axes[1].set_ylabel("Variance of tavg")
axes[1].set_title("    tavg ()")
axes[1].grid(True, linestyle='--', color='gray', alpha=0.7)
axes[1].legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
