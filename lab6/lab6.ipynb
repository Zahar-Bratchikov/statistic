```python
from scipy.stats import norm, t, chi2
import pandas as pd

```

# 1) С помощью пакета scipy.stats вычислите следующие вероятности:

```python
# Задача 1.1: Нормальное и t-распределения
# Данные:
confidence_level_a = 0.8
confidence_level_b = 0.4
confidence_level_c = 0.3
degrees_of_freedom_t = 5

# 1.1.1: Стандартное нормальное распределение (N(0,1))
# a. P(X ∈ [-a, a]) = 0.8
a_norm = norm.ppf(0.9, loc=0, scale=1)  # symmetrical interval for 80% probability

# b. P(X > b) = 0.4
b_norm = norm.ppf(1 - confidence_level_b)

# c. P(X < -c) + P(X > c) = 0.3
c_norm = norm.ppf(1 - confidence_level_c / 2)

# 1.1.2: T-распределение с 5 степенями свободы (T(5))
# a. P(X ∈ [-a, a]) = 0.8
a_t = t.ppf(0.9, degrees_of_freedom_t)

# b. P(X > b) = 0.4
b_t = t.ppf(1 - confidence_level_b, degrees_of_freedom_t)

# c. P(X < -c) + P(X > c) = 0.3
c_t = t.ppf(1 - confidence_level_c / 2, degrees_of_freedom_t)
print('1.1')
{
    'norm': {'a': a_norm, 'b': b_norm, 'c': c_norm},
    't': {'a': a_t, 'b': b_t, 'c': c_t}
}
```

```python
# Задача 1.2: Вероятности для нормального и t-распределений
# Нормальное распределение (N(0,1))
# γ1: P(X ∈ [-0.1, 0.1])
gamma1_norm = norm.cdf(0.1) - norm.cdf(-0.1)

# γ2: P(X > 0.7)
gamma2_norm = 1 - norm.cdf(0.7)

# γ3: P(X < -0.4) + P(X > 0.4)
gamma3_norm = norm.cdf(-0.4) + (1 - norm.cdf(0.4))

# T-распределение с 5 степенями свободы (T(5))
# γ1: P(X ∈ [-0.1, 0.1])
gamma1_t = t.cdf(0.1, degrees_of_freedom_t) - t.cdf(-0.1, degrees_of_freedom_t)

# γ2: P(X > 0.7)
gamma2_t = 1 - t.cdf(0.7, degrees_of_freedom_t)

# γ3: P(X < -0.4) + P(X > 0.4)
gamma3_t = t.cdf(-0.4, degrees_of_freedom_t) + (1 - t.cdf(0.4, degrees_of_freedom_t))
print('1.2')
{
    'norm': {'gamma1': gamma1_norm, 'gamma2': gamma2_norm, 'gamma3': gamma3_norm},
    't': {'gamma1': gamma1_t, 'gamma2': gamma2_t, 'gamma3': gamma3_t}
}
```

```python
# Задача 1.3: Хи-квадрат распределение с 10 степенями свободы (χ²(10))
# a. P(X > a) = 0.4
a_chi2 = chi2.ppf(1 - confidence_level_b, 10)

# b. P(X < b) = 0.2
b_chi2 = chi2.ppf(confidence_level_b / 2, 10)

# γ: P(X > 12)
gamma_chi2 = 1 - chi2.cdf(12, 10)
print('1.3')
{
    'chi2': {'a': a_chi2, 'b': b_chi2, 'gamma': gamma_chi2}
}
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


# вар 10
# (4 + 7 + 16) / 3 = 9
# 9 + 1 = 10
sigma = 1  # стандартное отклонение
mu = 10       
N = 50       
gamma = 96

sample = stats.norm.rvs(loc=mu, scale=sigma, size=N)

print(sample)
```

```python
gamma = 0.96
# Пункт (a): Доверительный интервал для среднего (σ известно)
z = norm.ppf((1 + gamma) / 2)
x_bar = np.mean(sample)  # выборочное среднее
epsilon_mean_known_sigma = z * sigma / np.sqrt(N)
ci_mean_known_sigma = (x_bar - epsilon_mean_known_sigma, x_bar + epsilon_mean_known_sigma)

# Пункт (b): Доверительный интервал для среднего (σ неизвестно)
s = np.std(sample, ddof=1)  # выборочное стандартное отклонение
t_val = t.ppf((1 + gamma) / 2, df=N-1)
ci_mean_unknown_sigma = (x_bar - t_val * s / np.sqrt(N), x_bar + t_val * s / np.sqrt(N))

# Пункт (c): Доверительный интервал для дисперсии (м.о. неизвестно)
chi2_lower = chi2.ppf((1 - gamma) / 2, df=N-1)
chi2_upper = chi2.ppf((1 + gamma) / 2, df=N-1)
ci_variance = ((N - 1) * s**2 / chi2_upper, (N - 1) * s**2 / chi2_lower)

# Вывод результатов
print("Выборочное среднее:", float(x_bar))
print("Половинная ширина интервала (σ известно):", float(epsilon_mean_known_sigma))
print("Доверительный интервал для среднего (σ известно):", float(ci_mean_known_sigma[0])), float(ci_mean_known_sigma[1])
print("Доверительный интервал для среднего (σ неизвестно):", float(ci_mean_unknown_sigma[0]), float(ci_mean_unknown_sigma[1]))
print("Доверительный интервал для дисперсии (м.о. неизвестно):", float(ci_variance[0]), float(ci_variance[1]))
```

```python
# Задание 4. Определение требуемого объёма выборки N1
epsilon_target = epsilon_mean_known_sigma / 2
N1_exact = (z * sigma / epsilon_target) ** 2
N1_min = int(np.ceil(N1_exact))

print("Текущее N:", N)
print("Необходимая новая ширина ε1 =", epsilon_target)
print("Расчётное значение N1:", N1_exact)
print("Минимальное целое N1 для выполнения условия:", N1_min)
print("Соотношение N1 / N:", N1_exact / N)

```

```python
df = pd.read_csv('daily_weather_data.csv')
display(df)
```

```python
from scipy import stats as st

# Задание 5.1: Фильтрация данных по странам и месяцу
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
selected_month = 6  # июнь
filtered_df = df[df['date'].dt.month == selected_month]
countries = filtered_df['country'].unique()[10:20]  # произвольно выбираем 10 стран
filtered_df = filtered_df[filtered_df['country'].isin(countries)]

filtered_df
```

```python
gamma = 0.95  # уровень доверия

list_statistics = (
    filtered_df.groupby("country")["tavg"]
    .agg(["mean", "std", "count"])
    .rename(columns={"mean": "mean_tavg", "std": "std_tavg"})
)

list_statistics = list_statistics[list_statistics["count"] > 1].copy()

# ----- Интервалы для среднего (t, γ) -----
list_statistics["t_multiplier"] = list_statistics["count"].apply(
    lambda n: st.t.ppf((1 + gamma) / 2, df=n - 1)
)
list_statistics["epsilon_mean"] = (
    list_statistics["t_multiplier"]
    * list_statistics["std_tavg"]
    / np.sqrt(list_statistics["count"])
)
list_statistics["ci_mean_lower"] = list_statistics["mean_tavg"] - list_statistics["epsilon_mean"]
list_statistics["ci_mean_upper"] = list_statistics["mean_tavg"] + list_statistics["epsilon_mean"]

# ----- Интервалы для дисперсии (χ², γ) -----
degrees_of_freedom = list_statistics["count"] - 1
list_statistics["variance_tavg"] = list_statistics["std_tavg"] ** 2

chi2_lower = chi2.ppf((1 - gamma) / 2, degrees_of_freedom)   # левый хвост
chi2_upper = chi2.ppf((1 + gamma) / 2, degrees_of_freedom)   # правый хвост

list_statistics["ci_var_lower"] = (
    degrees_of_freedom * list_statistics["variance_tavg"] / chi2_upper
)
list_statistics["ci_var_upper"] = (
    degrees_of_freedom * list_statistics["variance_tavg"] / chi2_lower
)

list_statistics[[
    "mean_tavg", "std_tavg", "count",
    "ci_mean_lower", "ci_mean_upper",
    "variance_tavg", "ci_var_lower", "ci_var_upper"
]]
```

```python

# Задание 5.3: Построение доверительных интервалов на графиках errorbar
fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

axes[0].errorbar(
    list_statistics.index,
    list_statistics['mean_tavg'],
    yerr=list_statistics['epsilon_mean'],
    fmt='o', capsize=6, label='95% CI for mean'
)
axes[0].set_ylabel("Mean tavg")
axes[0].set_title("Доверительные интервалы для среднего tavg (июнь)")
axes[0].grid(True, linestyle='--', color='gray')
axes[0].legend()

variance_yerr = np.vstack([
    list_statistics['variance_tavg'] - list_statistics['ci_var_lower'],
    list_statistics['ci_var_upper'] - list_statistics['variance_tavg']
])
axes[1].errorbar(
    list_statistics.index,
    list_statistics['variance_tavg'],
    yerr=variance_yerr,
    fmt='s', capsize=6, color='tab:red', label='95% CI for variance'
)
axes[1].set_xlabel("Country")
axes[1].set_ylabel("Variance of tavg")
axes[1].set_title("Доверительные интервалы для дисперсии tavg (июнь)")
axes[1].grid(True, linestyle='--', color='gray')
axes[1].legend()

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
